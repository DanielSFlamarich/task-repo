#### 2. 6. 3. Outliers

Most of the variables show significant skewness towards small values but a few extreme cases. Before addressing these asymmetries, one way to make sure we make transformations more effective and improve overall data quality is to handle these extreme values we are seeing. To treat these outliers with a robust method that is also interpretable and easy to comprehend, we'll use the **Tukey method**, which identifies outliers based on the *Interquartile Range* and a value we use as a multiplier to pinpoint a threshold from which **we'll label and cap outliers**. A small multiplier $\alpha$ will identify many outliers, at the risk of having lots of false positives, and a large multiplier will identify less outliers, but will also have the risk of another error, this time false negative error. We'll be using a conventional $\alpha$ of $1.5$ for the two types of attributes. The steps are:

1. **Calculate the Quartiles:**
   - *Q1 (First Quartile):* The 25th percentile of the data.
   - *Q3 (Third Quartile):* The 75th percentile of the data.

2. **Calculate the Interquartile Range (IQR):**
   
   $ \text{IQR} = Q3 - Q1 $

4. **Determine the Lower and Upper Boundaries:**
   - *Lower Boundary:*
     $ \text{Lower Boundary} = Q1 - \alpha \times \text{IQR} $
   - *Upper Boundary:*
     $ \text{Upper Boundary} = Q3 + \alpha \times \text{IQR} $
     

5. **Identify and Cap Outliers:**
   - Any data points below the lower boundary or above the upper boundary are considered outliers.
  
This method will preserve the data structure and its integrity. It also doesn not assume normality of the data, as parametric methods would do.

````python
for i in range(len(dfs)):
    dfs[i] = cap_outliers(dfs[i], int_cols)

# reassign back to original variables
df_tr, df_ts = dfs
````

#### 2. 6. 4. Transformations 

After handling extreme values, we can start scaling our data. Scaling aims to **reduce the skewness** we've seen in most of our attributes, **improves interpretability** of coefficients in regression scenarios, and **improves model performance**, as makes all features contribute more ewually to the learning process.

Here are a few thoughts on the options we have, from a normal *log-normalization* to more complex techniques like *Z-Score* normalization: 

- **Log**: Reduces positive skewness, handling a wide range of values. However, it's undefined for 0 or negative values and it doesn't normalize necessarily.
- **Square Root**: Effective for moderate skewness, but it's less effective with extreme skewness. 
- **Box-Cox**: Flexible for different types of skewness and also ideal for data that is strictly positive.
- **Normalization**: Scales to a unit scale range $[0,1]$, but highly sensitive to outliers and does not address skewness.
- **Standarization**: Centers data and normalizes it, but it cna't transform skewed data to normal.

For our case, both *Log transformation* and *Box-Cox* could be good candidates. We'll still aim for simplicity, so we'll **choose the log transformation**. Given that we have indeed values that are zero in different columns, we'll sum up 1 to all transformations while transforming.

````python
transform_cols = df_tr.drop(columns=['SeriousDlqin2yrs', 'MonthlyIncome']).columns
````

````python
# Apply log transformation to each dataset
for df in dfs:
    for col in transform_cols:
        df['Log' + col] = np.log1p(df[col])

df_tr, df_ts = dfs
````

#### 2. 6. 5 Correlations

Correlations are important because they can point to different kinds of relations in a dataset that may or may not be visible at first sight.

````python
# get log-transformed columns and target variable
log_transformed_cols = [col for col in df_tr.columns if col.startswith('Log')]
target_variable = 'SeriousDlqin2yrs'

# new DataFrame with only log-transformed columns and the target variable
df_tr_log_transformed = df_tr[log_transformed_cols + [target_variable]]
````

Correlation matrix shows correlations coefficient to measure the linear relationship between two variables. As per the method, we'll use Spearman's test, as it's a non-parametric test that captures simple relationships that are basically monotonic.
The values range from $[-1, 1]$ where:

- **r=1**: points to a perfect positive linear relationship.
- **r=0**: shows no linear relationship.
- **r=-1**: points to a perfect negative linear relationship.

````python
# find correlations
correlation_matrix = df_tr_log_transformed.corr(method='spearman')
````

````python
# skip those vars with not enough variability to appear in the matrix (those creating nan values)
correlation_matrix = correlation_matrix.dropna(axis=0, how='all').dropna(axis=1, how='all')

# visualize the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
````
