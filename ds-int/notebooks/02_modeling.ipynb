# 3. Modeling

The models chosen aim to balance interpretablility and simplicity (which is often the same thing), as explainability should not only be about perfoming techniques post-modeling, but it should be kept in mind in all steps. Using the proper algorithm for the proper task is advantegeous in XAI frameworks and mindsets. Aiming for this transparency, we opt for a statistical algorithm, a tree based one, and a boosting one.

- **Logistic Regression**: good base model, interpretable coefficients
- **Random Forest**: Non-linear relationships (we saw at the correlation part they might not be clear), also feature importance.
- **Gradient Boosting**: Given our dataset is not high dimensional, the sequential nature of GB usually helps reaching good accuracy (SVM are often used in those cases, it can capture non-linear relationships with the kernel trick).

### 3. 1. Define models and benchmark

We'll start with simple models so we have an average evaluation metric to serve as a benchmark. We'll use AUC Curve (Area Under Receiver Operating Characteristic Curve) which is widely used as a metric for evaluating binary classificators. Its characteristics make it the preferred option over other methods like precision-recall curve, which can be harder to interpret.

- AUC Curves are threshold independent: It plots the **True Positive rate** agains the **False Positive Rate** at various threshold settings. Evaluates model's performance across all classification thresholds.
- AUC Curves provides a single scalar value to summarize the performance, so it's easily interpretable.
- They can be interpreted as a probability, which also helps with clarity and it's extremely intuitive even for non-technical people.

Recall metric could also be used, as it describes the porportion of actual positives (failing scores) that are correctly identified. This means it ensures that most of the high-end borrowers are detected. In a financial institution, missing a high-risk borrower (False Negative) is more costly than incorrectly flagging a low-risk one (False Positive). More details on the evaluation metrics pros and cons are available in answer number 2 in ```03_questions.ipynb``` notebook.

````python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
}
````

At this step, the function ```evaluate_model``` we'll be using adds an important step to estimate how our model is learning. Cross validation ensures that the AUC score is used to calculate training and validating the model $n$ times on different splits of the data, providing a mean score reflecting the model's generalization capability. More specifically, data is divided into $k$ subsets, and model is trained on $k - 1$ folds and validated with the remaining one. We repeat the process $k$ times, with each folder used once as validation.
In essence, we cross validate so:

- We have a more accurate measure of the model performance compared to a single train-test split.
- It helps us use our data better.
- Reduces variance, so it helps with overfitting.

In our case, we've set the number of folders to 5, but trying different values can sometimes be a good practice. Our chosen $k$ though strikes a balance between variance (the error introduced by overcomplexity) and bias (the error introduced by oversimplification), and it's also efficient for our use case.

````python
%%time
model_results = {}
for name, model in models.items():
    print(f"Evaluating {name}...")
    mean_auc, std_auc = evaluate_model(model, X_train, y_train)
    model_results[name] = (mean_auc, std_auc)
````

````python
# results
for name, (mean_auc, std_auc) in model_results.items():
    print(f"{name} - Mean AUC: {mean_auc:.4f}, Std AUC: {std_auc:.4f}")
````

````python
# best model based on mean AUC
best_model_name = max(model_results, key=lambda k: model_results[k][0])
best_model = models[best_model_name]

print(f"Best model: {best_model_name}")
````

The model chosen is Gradient Boosting, which is a kind of **ensemble-type models**. Boosting techniques create stronger models by **combining the outputs of several weaker models**. Mainly, the idea is to focus on errors of previous classifiers and adjust the next ones whilst correcting them. *Gradient Boosting Machines*, specifically, work on the residual errors of previous trees sequentially, using gradient descent to minimize the loss function. In our case, a binary classification, loss function used is usually Binary Cross-Entropy.

$$
L(y, \hat{y}) = - \frac{1}{n} \sum_{i=1}^{n} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
$$

where:
- $ y_i $ is the true label for the \(i\)-th instance.
- $ \hat{y}_i$ is the predicted probability of the positive class for the \(i\)-th instance.
- $ n $ is the total number of instances.


### 3. 2. Train Best Benchmark model

````python
# train the best model on the full training data
best_model.fit(X_train, y_train)

# evaluate the best model on the validation set
y_val_pred = best_model.predict_proba(X_val)[:, 1]
val_auc = roc_auc_score(y_val, y_val_pred)
print(f"Validation AUC for the best model ({best_model_name}): {val_auc:.4f}")
````

````python
# ROC curve
fpr, tpr, thresholds = roc_curve(y_val, y_val_pred)

plt.plot(fpr, tpr, label=f"{best_model_name} (AUC = {val_auc:.4f})")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="best")
plt.show();
````

### 3. 3. Build Best model

Once we have a benchmark and we've decided which model to go for, we'll prepare a grid search to go through the hyperparameters and find the best ones. In the parameter grid, we can see Gradient Boosting most usual parameters:

- ```n_estimators```: Number of boosting stages  to be run, or trees in the ensemble. We've chosen a range that should help balance complexity and efficiency.
- ```max_depth```: Maximum depth of each tree. We try values to explore several levels of complexity in order to avoid under/overfitting.
- ```learning_rate```: The size of the step shrinkage to prevent overfitting. We try values to ensure the model can learn at different rates.

Grid Search is an important step in the process, as it'll define how we look for the best possible combination of parameters. We have two options:

- **GridSearch**: Exhaustive, ensures best possible combination within the specified grid. It's also deterministic, so it'll always provide the same result at the expense of computational time and resources.
- **RandomSearch**: More efficient as it evaluates a random subset of the space. It might miss the best possible combination (it's non-deterministic), but it's way faster. Again, for simplicity, we'll use a randomized search.

````python
%%time
if best_model_name == "Gradient Boosting":
    param_distributions = {
        'n_estimators': [100, 200],
        'max_depth': [3, 5, 7],
        'learning_rate': [0.01, 0.1, 0.2]
    }

    random_search = RandomizedSearchCV(estimator=GradientBoostingClassifier(random_state=42), 
                                       param_distributions=param_distributions, 
                                       n_iter=10,  # specify the number of parameter settings that are sampled
                                       cv=5,       # cross validation folders
                                       scoring='roc_auc', 
                                       random_state=42) # reproducibility
    random_search.fit(X_train, y_train)

    # print chosen hyperparameters
    best_model = random_search.best_estimator_
    print(f"Best hyperparameters for {best_model_name}: {random_search.best_params_}")

    # evaluate tuned model on the validation set
    y_val_pred_tuned = best_model.predict_proba(X_val)[:, 1]
    val_auc_tuned = roc_auc_score(y_val, y_val_pred_tuned)
    print(f"Validation AUC for the tuned best model ({best_model_name}): {val_auc_tuned:.4f}")

    # plot ROC curve for the tuned model
    fpr, tpr, thresholds = roc_curve(y_val, y_val_pred_tuned)
    plt.plot(fpr, tpr, label=f"{best_model_name} (Tuned AUC = {val_auc_tuned:.4f})")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend(loc="best")
    plt.show();
````

````python
# make predictions on the Test Set
y_test_pred = best_model.predict_proba(X_test)[:, 1]
````

```LogRevolvingUtilizationOfUnsecuredLines``` **(0.72)** has an overwhelmingly importance in this score, indicating it's the most influential factor predicting the target variable. Features between **0.05** to **0.08**  (```LogDebtRatio```, ```LogImputedMonthlyIncome```, ```Logage```, ```LogNumberOfOpenCreditLinesAndLoans```) would have moderate importance, as they contribute to the prediction but in to a much lesser extent compared to the main predictor.

We can consider the rest of the variables to have very low to negligible importance, given these results. Other things to consider:

- The model relies heavily on the ```LogRevolvingUtilizationOfUnsecuredLines``` feature, suggesting that **credit utilization is a crucial indicator of credit risk**.
- The next most important features, like debt ratio and monthly income, are also **logically consistent** with credit risk assessment, even with no further domain knowledge.
- Features with **zero importance could be candidates for removal**, simplifying the model and potentially **improving performance and interpretability**.

### 4. 2. SHAP

````python
explainer = shap.Explainer(best_model, X_train)

# calculate SHAP values for the validation set
shap_values = explainer(X_val, check_additivity=False)
````

````python
# extract SHAP values for the dependence plot
shap_values_array = shap_values.values
X_val_array = X_val.values
````

````python
# Summary plot
shap.summary_plot(shap_values, X_val)
````

The SHAP summary provides a clear and comprehensive view of the feature importance and the effects of individual features on the model's predictions. A more general discussion of how SHAP can help and works is available in question number 5 in the ```03_questions.ipynb``` notebook.

- ```LogRevolvingUtilizationOfUnsecuredLines``` shows its importance again, this time by showing that high values from it increase the model's predicted risk. Its widest spread also reinforces its dominant role.
- ```Logage``` appeared as a moderate contributor in the feature importance result, but SHAP analysis considers it a crucial factor. Lower values (younger ages) tend to increase risk prediction.
- On the other hand, ```LogImputedMonthlyIncome``` and ```LogNumberOfOpenCreditLinesAndLoans``` repeat result, being considered as moderate contributors, and the same is true for ```LogDebtRatio```
- Vertical dispersion can point to interactions between features. For instance, ```LogNumberRealEstateLoansOrLines```, which shows moderate contribution, could be associated with ```LogNumberOfOpenCreditLinesAndLoans``` and/or ```LogImputedMonthlyIncome```. We saw this correlations in the exploratory analysis (see ```section 2.5.5 ``` of the notebook ```01_data_prep_and_EDA.ipynb```)

To finish these insights, Let's see the individual plot for the most important attribute.

````python
# point wise for main feature
shap.dependence_plot('LogRevolvingUtilizationOfUnsecuredLines', shap_values_array, X_val_array, feature_names=features)
````

We see an expected positive correlation between ```LogRevolvingUtilizationOfUnsecuredLines``` and the SHAP value; as the variable increases, so does the risk. SHAP values increase quite rapidly from 0 to 1, suggesting even small increases in utilization can impact the model's prediction significantly. Another way of looking at this is that the model's prediction is very sensitive to the attribute's variability.
A non-linear pattern arises for higher values of the variable (6 to 10), but values are scattered.

By default, SHAP also selects what it determines to be the most significant interaction feature (in our case, ```LogNumberRealEstateLoansOrLines```). Color suggests that the impact of this feature might be influenced by the number of real state loans or lines. This confirms our correlation analysis in ```01_data_prep_and_EDA.ipynb-section 2.5.5 ``` .
