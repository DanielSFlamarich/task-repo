```python
import sys
import os

import pandas as pd
import numpy as np
import datetime as dt
from datetime import timedelta
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.insert(0, os.path.abspath('../'))
```

```python
# positions and sizes for XAI methods 
methods = {
    'Model Fitting': (5, 0, 0),  # vertical line for modeling
    'Feature Importance': (2.5, 2, 200),
    'PCA': (1.6, 1, 100),
    'PDP': (2.5, 3, 200),
    'Randomized Experiments': (1, 9, 200),
    'LIME': (5.5, 5, 100),
    'SHAP': (5.3, 6, 300),
    'Counterfactual Explanations': (5.2, 7, 100),
    'Causal Inference Methods': (7, 8.5, 200),
    'Attention Maps': (8, 8, 300),
    'OLS': (6, 2, 200),
    'Statistical Tests': (2.8, 4, 100),
}

# sort methods by size
sorted_methods = sorted(methods.items(), key=lambda item: item[1][2], reverse=True)


fig, ax = plt.subplots(figsize=(14, 8))

# plot each method with its respective coordinates and sizes
for method, (x, y, size) in methods.items():
    if method == 'Model Fitting':
        ax.axvline(x=x, color='gray', linestyle='--')
        ax.text(x, 0.5, 'Modeling', rotation=90, verticalalignment='center', fontsize=12, color='gray')
    else:
        ax.scatter(x, y, s=size, label=method, alpha=0.6, edgecolors="w", linewidth=2)
        ax.text(x, y, method, fontsize=12, ha='right')

ax.set_xlabel('Chronological Order (Scale is arbitrary)')
ax.set_ylabel('Explainability-Causality')
ax.set_title('Visual Summary of XAI Methods with Dataset Suitability')

# limits and grid
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.grid(True)

# sorted legend
handles, labels = ax.get_legend_handles_labels()
sorted_handles_labels = sorted(zip(handles, labels), key=lambda x: methods[x[1]][2], reverse=True)
sorted_handles, sorted_labels = zip(*sorted_handles_labels)
ax.legend(sorted_handles, sorted_labels, loc='upper left', bbox_to_anchor=(1, 1))
plt.show();
```
